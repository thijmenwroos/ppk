```python
sample_data = read_papyrus(is3d=False, plusplus = True, source_path='papyrus_data')
```

```python
from papyrus_scripts.reader import read_protein_set
protein_data = read_protein_set(source_path=None)
```

```python
from papyrus_scripts.preprocess import keep_quality, keep_protein_class, keep_type, keep_organism, keep_source
filter1_it = keep_quality(data=sample_data, min_quality='high')
print(len(filter1_it))
filter2_it = keep_protein_class(data=filter1_it, protein_data=protein_data, classes=[{'l2': 'Kinase'}])
filter4 = keep_organism(data=filter2_it, protein_data=protein_data, organism=['Human'], generic_regex=True)
print(len(filter2_it))
print(len(filter2_it[filter2_it['source']=='Christmann2016']))
print(len(filter2_it[filter2_it['source']!='Christmann2016']))
```
